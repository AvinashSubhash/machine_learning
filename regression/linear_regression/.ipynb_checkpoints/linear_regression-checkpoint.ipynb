```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def data_creator():
    data = np.ones((50,1))
    out = np.zeros(data.shape)
    for i in range(data.shape[0]):
        data[i] = random.uniform(1,1000)
        out[i] = 2*(data[i]) + 5 + random.randint(-350,350)

    #print(data,out)
    return data,out
        
```

```python
data,out = data_creator()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, out)
regress = LinearRegression()
regress.fit(X_train, y_train)
```

```python
print("Accuracy: ",int(regress.score(data,out)*100),"%")
```

```python
y_pred = regress.predict(data)
data = data.reshape((data.shape[1],data.shape[0]))
out = out.reshape(data.shape)
y_pred = y_pred.reshape(data.shape)
print("Predicted Output VS Actual Output\n")
plt.scatter(data,out,color='r')
plt.scatter(data,y_pred,color='b')
plt.plot(data,y_pred)
plt.show()
print("\n\nOutput Data distribution VS Prediction Data distribution\n")
sns.distplot(out,color='r')
sns.distplot(y_pred)
plt.show()
```
